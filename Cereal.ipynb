```python
plt.figure(figsize=(8,6))
plt.title("Overall Cereals Data Distribution")
sns.histplot(data = df)
```

```python
def plot_histogram(column_data, column_name):
    plt.figure(figsize=(5, 3))
    plt.title(f"Distribution of {column_name}")
    sns.histplot(column_data, kde=True) 
    plt.show()
```

```python
plot_histogram(df['mfr'], 'Manufacturer')
```

```python
plot_histogram(df['calories'], 'Calories')
```

```python
plot_histogram(df['fat'], 'Fat')
```

```python
plot_histogram(df['fiber'], 'Fiber')
```

```python
plot_histogram(df['sugars'], 'Sugar')
```

```python
plot_histogram(df['cups'], 'Cups')
```

```python
plot_histogram(df['rating'], 'Consumer Rating')
```

```python
df_sorted = df.sort_values(by='rating', ascending=False)

plt.figure(figsize=(14, 7))
plt.title("Brands and Their Ratings")
plt.xticks(rotation=90)
sns.barplot(data=df_sorted, x=df_sorted['name'], y=df_sorted['rating'])
```

```python
numeric_columns = df.drop(columns=['name', 'type', 'mfr','shelf', 'cups', 'weight'])


correlation_matrix = numeric_columns.corr(method = 'pearson')

# Create a heatmap for correlation with 'rating'
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix[['rating']].sort_values(by='rating', ascending=False), annot=True)
plt.title("Correlation Heatmap with Ratings")
plt.show()
```
