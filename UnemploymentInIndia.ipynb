```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dataset['Region'] = dataset['Region'].astype('category')
```

```python
fig = px.bar(dataset, x = "Region", y = " Estimated Unemployment Rate (%)", color = "Region", title = "Unemploment Rate in India")
fig.update_layout(xaxis = {'categoryorder':'total descending'})
fig.show()
```

```python
fig = px.bar(dataset, x = "Region", y = " Estimated Employed", color='Region', title = "Employed Rate in India")
fig.update_layout(xaxis = {'categoryorder':'total descending'})
fig.show()
```

```python
plt.figure(figsize=(14, 7))
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=dataset)
plt.xticks(rotation=90)
plt.title('Box Plot of Estimated Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')

plt.show()
```

```python
plt.figure(figsize=(10,5))
dataset.groupby('Area')[' Estimated Unemployment Rate (%)'].mean().plot(kind='pie', autopct='%1.1f%%')
plt.title('Area wise Unemployment Rate')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
corr = dataset[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']].corr()
sns.heatmap(corr, annot=True, linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
avg_unemployement_over_time = dataset.groupby(' Date')[' Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).reset_index()
avg_unemployement_over_region = dataset.groupby('Region')[' Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).reset_index()
avg_people_employed_over_region = dataset.groupby('Region')[' Estimated Employed'].mean().sort_values(ascending=False).reset_index()
avg_labour_part_rate_over_region = dataset.groupby('Region')[' Estimated Labour Participation Rate (%)'].mean().sort_values(ascending=False).reset_index()
avg_unemployement_area_wise = dataset.groupby('Area')[' Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False).reset_index()
highest_unemployement_rate_month = dataset.groupby(' Date')[' Estimated Unemployment Rate (%)'].max().sort_values(ascending=False).reset_index()

# print all the dataframes
print("Average Unemployment Rate over time\n", avg_unemployement_over_time)
print("\nAverage Unemployment Rate over Region\n", avg_unemployement_over_region)
print("\nAverage People employed over Region\n", avg_people_employed_over_region)
print("\nAverage Labour Participation Rate over Region\n", avg_labour_part_rate_over_region)
print("\nAverage Unemployment Rate over Area\n", avg_unemployement_area_wise)
print("\nHighest Unemployment Rate Month wise\n", highest_unemployement_rate_month)
```
